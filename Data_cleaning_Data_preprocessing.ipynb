```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path ='Titanic-Dataset.csv'
df=pd.read_csv(file_path)

df_cleaned = df.drop_duplicates()

df_cleaned['PassengerId'] = df_cleaned['PassengerId'].fillna(df_cleaned['PassengerId'].mean())
df_cleaned['Survived'] = df_cleaned['Survived'].fillna(df_cleaned['Survived'].mean())
df_cleaned['Pclass'] = df_cleaned['Pclass'].fillna(df_cleaned['Pclass'].mean())
df_cleaned['Name'] = df_cleaned['Name'].fillna(df_cleaned['Name'].mode()[0])
df_cleaned['Sex'] = df_cleaned['Sex'].fillna(df_cleaned['Sex'].mode()[0])
df_cleaned['Age'] = df_cleaned['Age'].fillna(df_cleaned['Age'].mean())
df_cleaned['SibSp'] = df_cleaned['SibSp'].fillna(df_cleaned['SibSp'].mean())
df_cleaned['Parch'] = df_cleaned['Parch'].fillna(df_cleaned['Parch'].mean())
df_cleaned['Ticket'] = df_cleaned['Ticket'].fillna(df_cleaned['Ticket'].mode()[0])
df_cleaned['Fare'] = df_cleaned['Fare'].fillna(df_cleaned['Fare'].mean())
df_cleaned['Cabin'] = df_cleaned['Cabin'].fillna(df_cleaned['Cabin'].mode()[0])
df_cleaned['Embarked'] = df_cleaned['Embarked'].fillna(df_cleaned['Embarked'].mode()[0])


#data visualization

ticket_counts = df_cleaned['Ticket'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(ticket_counts.index, ticket_counts.values)
plt.xlabel('Ticket')
plt.ylabel('Count')
plt.title('Ticket Distribution')
plt.xticks(rotation=90)
plt.show()


plt.figure(figsize=(10, 5))
df_cleaned['Age'].hist(bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()

plt.figure(figsize=(10, 5))
df_cleaned['Fare'].hist(bins=20, color='salmon', edgecolor='black')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Fare Distribution')
plt.show()

plt.figure(figsize=(10, 5))
df_cleaned['Pclass'].hist(bins=3, color='lightgreen', edgecolor='black')
plt.xlabel('Pclass')
plt.ylabel('Frequency')
plt.title('Pclass Distribution')
plt.show()

plt.figure(figsize=(10, 5))
df_cleaned['Survived'].hist(bins=2, color='lightcoral', edgecolor='black')
plt.xlabel('Survived')
plt.ylabel('Frequency')
plt.title('Survived Distribution')
plt.show()


```
